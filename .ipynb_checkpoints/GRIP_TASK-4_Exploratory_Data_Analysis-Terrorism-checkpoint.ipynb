```python
name_lis = list(df.columns[df.isnull().any()])
name_lis
```

```python
null = dict()
for i in name_lis:
    percent_null = df[i].isnull().sum()/len(df[i])*100
    if percent_null >= 70:
        null[i]=percent_null

null
```
